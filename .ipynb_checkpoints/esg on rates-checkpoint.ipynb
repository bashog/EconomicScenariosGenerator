# Importation of librairies and data

```python
# reload
%reload_ext autoreload
%autoreload 2

import pandas as pd
import random
import sys
sys.path.append('./')

from timeseries import TimeSeries
```

```python
dftaux = pd.read_csv(filepath_or_buffer = './datas/exchange_rates_ecb_eur_updated_2023.csv',sep = '\t')
dftaux.head()
```

# Initialisation of the multi timeseries

```python
symbols = ['USD', 'JPY', 'GBP', 'CAD']
```

```python
random.seed(14021999)
ts = TimeSeries(dftaux, symbols)
ts.pre_processing(method_return='arithmetic')
```

```python
ts.plot(type_plot='rates')
```

# Analysis of the timeseries

```python
ts.plot(type_plot='rates')
```

```python
ts.plot(type_plot='returns', frequency = 'weekly')
```

```python
ts.statistics()
```

```python
ts.correlation()
```

## Bootstrap method

```python
test_dates = '2021-09-04'
plot_from='2020-01-01'
scenarios = 100
frequency = 'annualy'
ts.bootstrap_esg(scenarios=scenarios, test_date=test_dates, plot_from=plot_from, frequency=frequency)
```

## RBM method

```python
test_dates = '2021-09-04'
plot_from='2020-01-01'
scenarios = 10
epochs = 5000
lr = 0.015
windows = 5
K = 100
frequency = 'weekly'
numba = True
ts.rbm_esg(scenarios, epochs, lr, K, numba, test_dates, plot_from, windows, frequency=frequency)
```

```python
import random
import numba
import numpy as np

def monte_carlo_pi(n_samples):
    acc = 0
    for i in range(n_samples):
        x = random.random()
        y = random.random()
        if x**2 + y**2 <= 1:
            acc += 1
    return 4*acc/n_samples

def sigmoid(x):
    return 1/(1+np.exp(-x))

@numba.njit
def monte_carlo_pi_numba(n_samples):
    acc = 0
    for i in range(n_samples):
        x = random.random()
        y = random.random()
        if x**2 + y**2 <= 1:
            acc += 1
    return 4*acc/n_samples
```

```python
monte_carlo_pi(10000000)
```
